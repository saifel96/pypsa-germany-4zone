```python
from entsoe import EntsoePandasClient
import pandas as pd
import os 
from dotenv import load_dotenv

load_dotenv()
ENTSOE_API_KEY = os.getenv('ENTSOE_API_KEY')
start_date = "20250101"
end_date = "20251231"
time_zone = "Europe/Berlin"
country_code = "DE_LU"

client = EntsoePandasClient(api_key=ENTSOE_API_KEY)

start = pd.Timestamp(start_date, tz=time_zone)
end = pd.Timestamp( end_date, tz=time_zone)
country_code = country_code

load = client.query_load(country_code,
                        start=start,
                        end=end)

load.to_csv('data/load_entsoe_data.csv')
```

```python
import pandas as pd

# Lade die Daten direkt
load = pd.read_csv("data/load_entsoe_data.csv",
    index_col=0,
    parse_dates=True
)

# Konvertiere zu datetime mit UTC
load.index = pd.to_datetime(load.index, utc=True)

# Optional: Timezone entfernen für einfachere Handhabung
load.index = load.index.tz_localize(None)

load_hourly = load.resample("1h").mean()

load_hourly.columns = ['Load_MW']

load_hourly.to_csv("data/load_hourly.csv")
load_hourly.columns = ['Load_MW']

load_hourly.to_csv("data/load_hourly.csv")

print("✓ Load hourly shape:", load_hourly.shape)
print(load_hourly.head())
```
